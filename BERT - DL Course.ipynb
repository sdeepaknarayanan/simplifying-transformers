## Train

```python
train_content = []
f = open("wiki.train.tokens", "r")

for line in f:
    train_content.append(line)

modified_train_content = []
for ix, line in enumerate(train_content):
    if '=' in line:
        continue
    elif len(line)<10:
        continue
    else:
        arr = line.strip('\n').strip(' ').split('.')
        elems_to_pop = []
        for i in range(len(arr)):
            if len(arr[i])<=3:
                elems_to_pop.append(arr[i])
        for elem in elems_to_pop:
            arr.remove(elem)
        count = 0
        buffer = ''
        for i in range(len(arr)):
            if buffer is '':
                try:
                    modified_train_content.append(arr[count]+' \t ' + arr[count+1] + '\n')
                    print(count)
                except:
                    if count == len(arr)-1:
                        buffer = arr[count]
            else:
                modified_train_content.append(buffer + ' \t ' + arr[count] + '\n')
                buffer = ''
            count = count + 2
            if count >= len(arr):
                break
```
